```python
import numpy as np
import matplotlib.pyplot as plt

title = ['breast-cancer', 'diabetes', 'digit', 'iris', 'wine']
train_accu = np.array([[97.44,97.07,97.07,98.17], [76.42,76.10,76.25,77.72], 
                      [85.25,98.18,62.25,100.0], [100.0,100.0,100.0,100.0], 
                      [62.68,59.86,94.37,97.89]])
test_accu = np.array([[97.06,97.06,97.06,96.32], [77.12,79.74,78.43,78.43], 
                       [83.5,95.0,62.0,100.0], [100.0,100.0,100.0,100.0], 
                       [61.0,61.1,94.44,77.78]])
labels = [0, 1, 2, 3] # on training instances or testing instances
models = [r'LR', r'NN', r'SVM_L', r'SVM_R']

plot_args = [{'c': 'b'},
             {'c': 'g'},
             {'c': 'r'},
             {'c': 'c'},
             {'c': 'y'}]

fig, axes = plt.subplots(2, 5, figsize=(25, 7))

for i in range(0, 5):
    ax1 = plt.subplot(2, 5, i+1)
    for a, label, args in zip(train_accu[i,:], labels, plot_args):
        ax1.plot([label], [a], 'o', **args)
    fig.legend(ax1.get_lines(), models, ncol=4, loc="upper center")        
    
    plt.title(title[i]+'(Training Instances)')
    plt.ylabel('Classification Accuracy') 
    plt.xlim((-1,5))
    plt.xticks(labels, models)
    plt.ylim((50, 110))     
        
    ax1 = plt.subplot(2, 5, 5+i+1)
    for a, label, args in zip(test_accu[i,:], labels, plot_args):
        ax1.plot([label], [a], 'o', **args)
    fig.legend(ax1.get_lines(), models, ncol=4, loc="upper center")        
        
    plt.title(title[i]+'(Testing Instances)')        
    
    plt.ylabel('Classification Accuracy') 
    plt.xlim((-1,4))
    plt.xticks(labels, models)
    plt.ylim((50, 110))       
    
plt.savefig('comparsion_accuracy.eps', dpi=200)
plt.show()
```
